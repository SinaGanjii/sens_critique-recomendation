```python
from bs4 import BeautifulSoup

def clean_html(text):
    if not isinstance(text, str):
        return ""
    
    return BeautifulSoup(text, "html.parser").get_text(separator=" ")

data["review_content"] = data["review_content"].apply(clean_html)
```

```python
from sentence_transformers import SentenceTransformer
```

```python
# la on cree le colom embeding pour que je puisse faire la embeding de chaque review
# only add colom ebdenif its all 
data["review_embedding"] = None
data.head()
```

```python
# la on load le model et et apres on vas calculer le embeding de chaque review et de le mettre dans la colom review_embedding
model = SentenceTransformer("paraphrase-multilingual-MiniLM-L12-v2")
```

```python
data["review_embedding"] = data["review_content"].apply(lambda x: model.encode(x))
```

```python
data.head()
# la tou va bein colum est bein et on vas just tester et mette un fontion pour trouver les top 5
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

print(f"critique 1 : {data['review_content'][0]}")
print(f"critique 2 : {data['review_content'][1]}")

embedding1 = data['review_embedding'][0]
embedding2 = data['review_embedding'][1]

similarity = cosine_similarity([embedding1], [embedding2])[0][0]

print(f"\nSimilarité entre les deux critiques: {similarity:.4f}")
```

```python
def find_top_5_similar_reviews(review_index, top_k=5):
    review_embedding = data.iloc[review_index]['review_embedding']

    similarities = []
    for i, embedding in enumerate(data['review_embedding']):
        if i != review_index:
            similarity = cosine_similarity([review_embedding], [embedding])[0][0]
            similarities.append((i, float(similarity)))

    top_5 = sorted(similarities, key=lambda x: x[1], reverse=True)[:top_k]
    for i, score in top_5:
        print(f"Critique {i} : {score:.4f}")
    
    
find_top_5_similar_reviews(2, 5)
```
